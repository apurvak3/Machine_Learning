```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
centroids =[(-5,5) ,(5,5)]
clusters_std =[1, 1]
```

```python
X, y = make_blobs(n_samples=100, centers=centroids, cluster_std=clusters_std, random_state=42)
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.show()
```
