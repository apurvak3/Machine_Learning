```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to convert boolean columns to integers (0 or 1)
def convert_bool_to_int_columns(df):
    """
    Converts all boolean columns in a DataFrame to integer type (0 for False, 1 for True).
    """
    for col in df.select_dtypes(include='bool').columns:
        df[col] = df[col].astype(int)
    return df
```
